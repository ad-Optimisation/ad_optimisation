```python
import pandas as pd
```

```python
color_data = pd.read_csv("../observations/color_feature.csv")
color_data
```

```python
# printing the shape of the data
print(f'shape: {color_data.shape}, size: {color_data.size}')
```

```python
# using duplicated method to check if there are any duplicates
color_data.duplicated().value_counts()
```

```python
# unique value counts
unique_counts = color_data['Assets'].nunique()
unique_counts
```

```python
counts_df = color_data['rgb'].value_counts()
counts_df
```

```python
color_data['r'] = color_data['rgb'].apply(lambda x: str(x).split(',')[0].removeprefix('('))
color_data['g'] = color_data['rgb'].apply(lambda x: str(x).split(',')[1])
color_data['b'] = color_data['rgb'].apply(lambda x: str(x).split(',')[2].removesuffix(')'))
```
